```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.xlabel('Grades')
plt.ylabel('Hours of Study')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
x = data[['Hours_of_Study']]
y = data[['Grades']]
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_(0)

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.plot(x,model.predict(x), color='yellow')
plt.xlabel('Grades')
plt.ylabel('Hours of Study')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
new_X = [[np.random.randint(100,200)]]
predicted_y = model.predict(new_X)
print('Value of X: ', new_X)
print('Predicted Value of Y: ', predicted_y)
```
