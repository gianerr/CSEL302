1. Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('hardwareStore.csv')
data.head()
```

2. Initial Data Analysis

```python
hardware = data.groupby('CATEGORY_NAME')['QUANTITY'].sum()

standard_cost_mean = data['STANDARD_COST'].mean()
standard_cost_median = data['STANDARD_COST'].median()
standard_cost_std = data['STANDARD_COST'].std()
print("STANDARD COST")
print("MEAN = " , standard_cost_mean)
print("MEDIAN = " , standard_cost_median)
print("SD = " , standard_cost_std)
print(" ")
list_price_mean = data['LIST_PRICE'].mean()
list_price_median = data['LIST_PRICE'].median()
list_price_std = data['LIST_PRICE'].std()
print("LIST PRICE")
print("MEAN = " , list_price_mean)
print("MEDIAN = " , list_price_median)
print("SD = " , list_price_std)
print(" ")
quantity_mean = data['QUANTITY'].mean()
quantity_median = data['QUANTITY'].median()
quantity_std = data['QUANTITY'].std()
print("QUANTITY")
print("MEAN = " , quantity_mean)
print("MEDIAN = " , quantity_median)
print("SD = " , quantity_std)
print(" ")
unique = data.groupby('PRODUCT_ID')['PRODUCT_ID'].count()
print(unique)
```

3. Category Analysis

```python
hardware.plot(kind = 'bar')
plt.xlabel('')
plt.ylabel('Number of Products')
plt.title('Products')
plt.xticks(rotation = 0)
plt.show
```

4. Cost and Price Analysis

```python
standard_cost_average = data['STANDARD_COST'].mean()
list_price_average = data['LIST_PRICE'].mean()

highest_standard_cost =  data.loc[data['STANDARD_COST'].idxmax(), 'PRODUCT_NAME']
highest_list_price = data.loc[data['LIST_PRICE'].idxmax(), 'PRODUCT_NAME']

print("The average of standard cost is = " , standard_cost_average)
print("The average of list price is =  " , list_price_average)
print("The product with the highest standard cost is = " , highest_standard_cost)
print("The product with the highest list price is = " , highest_list_price)
```

5. Location Analysis

```python
location = data.groupby('STATE')['QUANTITY'].sum()
highest_location = location.idxmax()
print("The location with the highest quantity of products in stock is in" , highest_location)
```
