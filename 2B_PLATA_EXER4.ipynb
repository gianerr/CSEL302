```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')
data.head()
```

```python
plt.scatter(data['Hours_of_Study'], data['Attendance_Percentage'], data['Travel_Time_to_School'], data['Grades'])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
X = data.drop('Grades', axis=1)
Y = data[['Grades']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
slope = model.coef_
intercept = model.intercept_
print('Slope:', slope)
print('Intercept:', intercept)
```

```python
new_data = np.array([[2,70,85,1,50]])
prediction = model.predict(new_data)
print(f"predict grades: {prediction[0]}")
```
